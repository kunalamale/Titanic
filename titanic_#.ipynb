# Missing Values

```python
df['Age']=df['Age'].fillna(method='pad')
```

```python
df['Age']
```

```python
df.isnull().sum()
```

```python
df['Embarked'].value_counts()
```

```python
df['Embarked']=df['Embarked'].fillna('S')
```

```python
df.drop(['Cabin'],axis=1,inplace=True)
```

```python
df
```

```python
df.drop(['PassengerId','Name','Ticket'],axis=1,inplace=True)
```

```python
df
```

```python
dummy=pd.get_dummies(df['Embarked'],drop_first=True)
dummy
```

```python
dummy1=pd.get_dummies(df['Sex'],drop_first=True)
dummy1
```

```python
df=pd.concat([df,dummy1,dummy],axis=1)
```

```python
df
```

```python
sns.countplot(data=df,x='Survived',hue='Embarked');
```

```python
sns.countplot(data=df,x='Survived',hue='Sex');
```

```python
sns.countplot(data=df,x='Survived',hue='Pclass',palette='magma');
```
